# Series de tiempo

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter
```

```python
data = pd.read_excel("/content/TCM.xlsx")
data
```

```python
#Crear Gráfico de líneas
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(data['Fecha (dd/mm/aaaa)'], data['Tasa de cambio representativa del mercado (TRM)'], color = 'green')

plt.xticks(rotation=45)

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Fecha (dd/mm/aaaa)')
plt.ylabel('Tasa de cambio representativa del mercado (TRM)')

plt.show()
```

```python
dates = pd.date_range(start='2023-01-01', periods=12, freq='ME')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame=({'Fecha (dd/mm/aaaa)': dates, 'Tasa de cambio representativa del mercado (TRM)': sales})

plt.plot(sales_data['Fecha (dd/mm/aaaa)'],sales_data['Tasa de cambio representativa del mercado (TRM)'], marker='o',
         linestyle = '--', label = 'Tasa de cambio')

plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

plt.xticks(rotation=45)

plt.title('Análisis Tasa de cambio')
plt.xlabel('Fecha (dd/mm/aaaa)')
plt.ylabel('Tasa de cambio representativa del mercado (TRM)')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
# Hacer media móvil para n = 3
n = 3
k = len(data['Tasa de cambio representativa del mercado (TRM)'])
mmovil = np.empty(k)
mmovil[:] = np.nan
ini = 0
for i in range(n-1,k):
  mmovil[i] = np.sum(data['Tasa de cambio representativa del mercado (TRM)'][ini:i+1])/n
  ini += 1
mmovil
```

```python
plt.plot(data['Fecha (dd/mm/aaaa)'], data['Tasa de cambio representativa del mercado (TRM)'], '--', label='Tasa de cambio representativa del mercado (TRM)')
plt.plot(data['Fecha (dd/mm/aaaa)'],mmovil, label='Media móvil n =3')
plt.title('Tasa de cambio por año')
plt.xlabel('Año')
plt.ylabel('Tasa de cambio')
plt.legend()
plt.show()
```
